# Comparing Performance of all 4 Learners
- Generating Plot for Average MSE for all learners

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(13, 4))
x = ["Default Xgboost", "Tuned RF", "Default RF", "Tuned Xgboost", "LGBM Default", "Default HGBT", "Tuned HGBT", "Tuned LGBM"]
Avg_MSE = [3.98, 3.92, 3.84, 3.79, 3.75, 3.74, 3.70, 3.67]
plt.plot(x, Avg_MSE, color='purple', linewidth=3)

# Adding Values
for i, j in zip(x, Avg_MSE):
    plt.text(i, j + 0.003, str(j)) # Add text slightly above the point

# Adding Title
plt.title("Model Selection: Avg MSE Scores, by Model")
plt.ylabel("Avg MSE")

# Show the plot
plt.show()
```

- Generating Plot for average r2 score, all learners

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(13, 4))
x = ["Default Xgboost", "Tuned RF", "Default RF", "Tuned Xgboost", "LGBM Default", "Default HGBT", "Tuned HGBT", "Tuned LGBM"]
Avg_r2 = [0.53, 0.54, 0.55, 0.55, 0.56, 0.56, 0.565, 0.57]
plt.plot(x, Avg_r2, color='brown', linewidth=3)

# Adding Values
for i, j in zip(x, Avg_r2):
    plt.text(i, j + 0.002, str(j)) # Add text slightly above the point

# Adding Title
plt.title("Model Selection: Avg r2 Scores, by Model")
plt.ylabel("Avg r2")

# Show the plot
plt.show()
```
